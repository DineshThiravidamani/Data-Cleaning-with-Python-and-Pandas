```python
import pandas as pd
```

```python
df=pd.read_excel('Customer Call List.xlsx')
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
df.drop(columns='Not_Useful_Column',inplace=True)
```

```python
df
```

```python
df['Last_Name']=df['Last_Name'].str.lstrip('/')
df['Last_Name']=df['Last_Name'].str.rstrip('_')
df['Last_Name']=df['Last_Name'].str.lstrip('...')
```

```python
df['Last_Name'].str.strip('/._')
```

```python
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
df
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x : str(x))
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x : x[0:3]+'-'+x[3:6]+'-'+x[6:10])
```

```python
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df['Phone_Number']=df['Phone_Number'].str.replace('Na--','')
```

```python
df
```

```python
df[['Street','State','Zipcode']]=df['Address'].str.split(',',expand=True)
```

```python
df
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Noo','No')
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('No/a','N/A')
```

```python
df
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].replace('N','No')
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].replace('Y','Yes')
```

```python
df
```

```python
df=df.fillna('')
```

```python
df
```

```python
for x in df.index:
  if df.loc[x,'Do_Not_Contact'] == 'Yes':
    df.drop(x,inplace = True)
```

```python
df
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].replace('','No')
```

```python
df
```

```python
df=df.reset_index(drop=True)
```

```python
df
```

```python
for x in df.index:
  if df.loc[x,'Phone_Number'] == '':
    df.drop(x,inplace=True)
```

```python
df
```

```python
df=df.reset_index(drop=True)
```

```python
df=df.drop(columns='Zipcode')
```

```python
df
```

```python
df.to_csv('After Data Cleaning.xlsx',index=False)
```
